```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

```

```python
normal  = np.random.rand(12,14)
```

```python
sns.heatmap(normal)
```

```python
sns.heatmap(normal,annot =True)
```

```python
sns.heatmap(normal,annot=True,vmin=0,vmax=1)
```

```python
sns.heatmap(normal,annot=True,vmin=0,vmax=3)
```

```python
sns.heatmap(normal,annot =True,cbar =False)
```

```python
sns.heatmap(normal,annot=True,cmap='RdBu')
```

```python
sns.heatmap(normal,annot =True,cmap='summer',cbar=False,linewidth=5)
```

```python
sns.heatmap(normal,annot=True,cbar=False,cmap='summer')
```
