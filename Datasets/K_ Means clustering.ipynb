```python
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('Mall_Customers.csv')
```

```python
X = data.iloc[:,[3,4]].values
```

```python
from sklearn.cluster import KMeans
wcss = []
for i in range(1,11):
    kmeans = KMeans(n_clusters=i,init='k-means++',max_iter=300,n_init=10,random_state=0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss)
```

```python
kmeans=KMeans(n_clusters= 5,init='k-means++',max_iter=300,n_init=10,random_state=0)
kmeans
```

```python
y_means =kmeans.fit_predict(X)
```

```python
plt.scatter(X[y_means==0,0],X[y_means==0,1],s=50,c='r',label='cluster 1')
plt.scatter(X[y_means==1,0],X[y_means==1,1],s=50,c='b',label='cluster 1')
plt.scatter(X[y_means==2,0],X[y_means==2,1],s=50,c='y',label='cluster 1')
plt.scatter(X[y_means==3,0],X[y_means==3,1],s=50,c='k',label='cluster 1')
plt.scatter(X[y_means==4,0],X[y_means==4,1],s=50,c='m',label='cluster 1')
plt.scatter(kmeans.cluster_centers_[:, 0],kmeans.cluster_centers_[:, 1],s=200,c='g',label ='centroid')
plt.legend()
```
